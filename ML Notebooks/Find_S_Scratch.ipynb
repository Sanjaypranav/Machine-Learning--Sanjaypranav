```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/drive/MyDrive/Benkyou/datasets ML/dataset_find_S.csv",sep="\t")
```

```python
d= np.array(data)[:,:-1]
target = np.array(data)[:,-1]
```

```python
def Train(c,t):

  for i, val in enumerate(t):
    if val == "Positive":
      hypothesis = c[i].copy()
      break
    
  for i,val in enumerate(c):
    if t[i] == "Positive":
      for x in range(len(hypothesis)):
        if val[x] == hypothesis[x]:
          hypothesis[x] == '?'
        else:
          pass
  return hypothesis
```

```python
print('Last hypothesis',Train(d,target))
```

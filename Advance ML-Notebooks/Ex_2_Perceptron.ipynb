###OR **GATE**

```python
def s_f(x):
  if x<0:
    return -1
  else:
    return 1

l_rate = 0.1
```

```python
Or=[]
for i in range(4):
  Or.append([[int(input('Enter X1=')),int(input('Enter x2=')),int(input('Enter bias='))],int(input('Enter Y ='))])
# data = np.array(data)
print(Or)
```

```python
weights = random.rand(3)
error =[]
n = 15
```

```python
weights
```

```python
for j in range(n):
  for x,target in Or:
    result = dot(weights,x)
    err = target - s_f(result)
    weights += l_rate * dot(err,x)
```
